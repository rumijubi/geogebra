```python
import matplotlib.pyplot as plt
import numpy as np
primary_color="#0da2cc"
secondary_color="#6eb819"
tertiary_color="#fd602e"
gray_color="#7f7f7f"
black_color="#4c4f54"
white_color="#ffffff"
color_list=[gray_color, primary_color,secondary_color,tertiary_color,black_color,white_color]
color_numa=0
color_numb=1
color_numc=2
color_numd=3
color_nume=0
lwl=2.0
x_s=-7
x_e=3
y_s=-255
y_e=200
lwa=0.6
# size of the figure
plt.figure(figsize=(5,5))

#axes
plt.axes([0,0,1,1], frameon=False)

#limit the axis
plt.xlim([x_s,x_e])
plt.ylim([y_s,y_e])
# Even though our axes (plot region) are set to cover the whole image with [0,0,1,1],
# by default they leave padding between the plotted data and the frame. We use tigher=True
# to make sure the data gets scaled to the full extents of the axes.
#plt.autoscale(tight=True)

#ploting axis
plt.grid(color='#c0c0c0',linestyle='dashed')
plt.plot([x_s,x_e],[0,0],color="#4c4f54",linewidth=lwa)
plt.plot([0,0],[y_s,y_e],color=color_list[0],linewidth=lwa)


x=np.linspace(x_s,x_e,100)
#ploting function
a=100*np.cos(x)
plt.plot(x,a,color=color_list[color_numa],linewidth=lwl)

#ploting function
b=100*np.sin(x)
plt.plot(x,b,color=color_list[color_numb],linewidth=lwl)
'''
#ploting function
c=100*np.tan(x)
plt.plot(x,c,color=color_list[color_numc],linewidth=lwl)

#ploting function
d=100/(np.sin(x))
plt.plot(x,d,color=color_list[color_numd],linewidth=lwl)

#ploting function
e=100/np.cos(x)
plt.plot(x,e,color=color_list[color_nume],linewidth=lwl)
'''







plt.savefig('k.png')
plt.show()
```
