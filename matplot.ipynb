```python
import matplotlib.pyplot as plt
import numpy as np

#color variable
gray_color="#7f7f7f"
primary_color="#0da2cc"
secondary_color="#6eb819"
tertiary_color="#fd602e"
black_color="#4c4f54"
white_color="#ffffff"

#linestyle
solid='solid'
ldashed=(0,(5,10))
dashed='dashed'

## notice the index of linestyle
ls_list=[solid,ldashed,dashed]

#### 2 change the linestyle ####
l_a=0
l_b=0
l_c=2
l_d=2
l_e=2

## notice the index of color
color_list=[gray_color, primary_color,secondary_color,tertiary_color,black_color,white_color]

#### 2 change the color ####
c_a=2
c_b=1
c_c=1
c_d=3
c_e=0

#### 1 chage the variable to limit the axis ####
x_s=-7
x_e=3
y_s=-255
y_e=200

#line-width of axis and lines
lwa=1
lwl=3.3

# size of the figure
plt.figure(figsize=(5,5))

#axes
plt.axes([0,0,1,1], frameon=False)

#limit the axis
plt.xlim([x_s,x_e])
plt.ylim([y_s,y_e])

#plt.autoscale(tight=True)

#ploting axis
plt.grid(color='#c0c0c0',linestyle=ls_list[1])
plt.plot([x_s,x_e],[0,0],color="#4c4f54",linewidth=lwa)
plt.plot([0,0],[y_s,y_e],color=color_list[0],linewidth=lwa)

# x value
x=np.linspace(x_s,x_e,100)

#### 0 change ploting function ####

#ploting function
a=100*np.cos(x)
plt.plot(x,a,color=color_list[c_a],linewidth=lwl,linestyle=ls_list[l_a],alpha=.7)

#ploting function
b=100*np.sin(x)
plt.plot(x,b,color=color_list[c_b],linewidth=lwl,linestyle=ls_list[l_b],alpha=.7)

#ploting function
c=100*np.tan(x)
plt.plot(x,c,color=color_list[c_c],linewidth=lwl,linestyle=ls_list[l_c],alpha=.7)

#ploting function
d=100/(np.sin(x))
plt.plot(x,d,color=color_list[c_d],linewidth=lwl,linestyle=ls_list[l_d],alpha=.7)

#ploting function
e=100/np.cos(x)
plt.plot(x,e,color=color_list[c_e],linewidth=lwl,linestyle=ls_list[l_e],alpha=.7)


plt.savefig('k.png')
plt.show()
```
